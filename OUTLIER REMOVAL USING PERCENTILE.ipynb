# Outlier detection and removal using percentile 

```python
import pandas as pd
import numpy as np

df=pd.read_csv('heights.csv')
df.head()
```

Knowing how many vaulue or percentage of data lie below 95 percetile

```python
max_threshold=df['height'].quantile(0.95)
max_threshold
```

Identifying an outlier (MAXIMUM VALUE)

```python
df[df['height']>max_threshold]
#The height cannot be 14 feet and hence an outlier.
```

INDENTIFYING AN OUTLIER FOR A MINIMUM VALUE

```python
min_threshold=df['height'].quantile(0.05)
min_threshold
```

```python
df[df['height']<min_threshold]
#The minimum of an adult cannot be 1.2 feet.
```

REMOVING OUTLIERS FROM THE DATASET

```python
# WE KEEP THE DATA IF IT IS GREATER THAN THE MIN THRESHOLD AND LESS THAN THE MAX THRESHOLD.
df[(df['height']>min_threshold)&(df['height']<max_threshold)]
```

# OUTLIER DETECTION AND REMOVAL FOR COMPLEX DATASET

```python
df=pd.read_csv('bhp.csv')
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
min_threshold=df['price_per_sqft'].quantile(0.001)
max_threshold=df['price_per_sqft'].quantile(.999)
print("The minimum threshold is:",min_threshold)
print("The maximum threshold is:",max_threshold)
```

FINDNG OUTLIER THAT IS LESS THAN THAN THE MIN_THRESHOLD

```python
df[df['price_per_sqft']<min_threshold]
```

FINDNG OUTLIER THAT IS GREATER THAN THAN THE MAX_THRESHOLD

```python
df[df['price_per_sqft']>max_threshold]
```

REMOVING THE OUTLIERS AND STORING THEM IN NEW DF

```python
df1=df[(df['price_per_sqft']>min_threshold)&(df['price_per_sqft']<max_threshold)]
df1
```

```python
 df1.describe()
```
